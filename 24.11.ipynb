```python
e = [4,7,2,9,12]
def array(arr):
    result_arr = []
    for i in arr:
        result_arr.append(i*i)
    return result_arr
r = array(e)
print(r)
```

```python
e = [-3,-2,-1,0,1,2,3]
a = int(input("Введите коэфицент а"))
b = int(input("Введите коэфицент b"))
c = int(input("Введите коэфицент c"))
def parabola(arr):
    res = []
    for i in arr:
        res.append(a*i**2 + b*i + c)
    return res
o = parabola(e)
print(o)

    
```
