```python
import tkinter as tk
root = tk.Tk()
root.mainloop()
```

```python
from tkinter import *
w = Tk()
w.title('')

la = Label(w,text=' ')
la.pack()

en = Entry(w,width=30)
en.pack()

bu = Button(w,text='  ')
bu.pack()

arr = ['1','2','3']
li = Listbox(w)
for i in arr:
    li.insert(END,i)
li.pack()

ch = BooleanVar()
chb = Checkbutton(w,text=',  ',variable=ch)
chb.pack()

te= Text(w,height=10,width=10)
te.pack()

sc = Scrollbar(w,command=te.yview)
sc.pack(side=RIGHT,fill=Y)
te.config(yscrollcommand=Scrollbar.set)

w.mainloop()
```

```python
q = Tk()

def gn():
    n = iu.get()
    gu.config(text=(':',n))

nu = Label(q,text='?')
nu.grid(row=0,column=0)
iu = Entry(q)
iu.grid(row=0,column=1)

su = Button(q,text='',command = gn)
su.grid(row=1,column=0)

gu = Label(q,text=' :')
gu.grid(row=1,column=1)

q.mainloop()
```
