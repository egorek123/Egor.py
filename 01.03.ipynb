```python
from tkinter import *
w = Tk()
w.title('Лох')

la = Label(w,text='Лох Лох')
la.pack()

en = Entry(w,width=30)
en.pack()

bu = Button(w,text='Лох Лох Лох')
bu.pack()

arr = ['1Лох','2Лох','3Лох']
li = Listbox(w)
for i in arr:
    li.insert(END,i)
li.pack()

ch = BooleanVar()
chb = Checkbutton(w,text='Да, Я лох',variable=ch)
chb.pack()

te= Text(w,height=10,width=10)
te.pack()

sc = Scrollbar(w,command=te.yview)
sc.pack(side=RIGHT,fill=Y)
te.config(yscrollcommand=Scrollbar.set)

w.mainloop()
```
