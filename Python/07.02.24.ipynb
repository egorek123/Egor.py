```python
import numpy as np
def f(x):
    return x**2 + 10*np.sin(x)

def grad_f(x):
    return 2*x + 10*np.cos(x)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-10,10,1000)
f_x = f(x)
f_grad = grad_f(x)

plt.scatter(x,f_x)
plt.scatter(x,f_grad)

plt.show()
```

```python
def grad_descent(start, learning_rate, n_iter):
    x = start
    trajectory = [x]
    for i in range(n_iter):
        x = x - learning_rate*grad_f(x)
        trajectory.append(x)
    return np.array(trajectory)
```

```python
start  = -5
learning_rate = 0.1
n_iter = 10

trajectory = grad_descent(start, learning_rate, n_iter)
print(trajectory)

print('Точка х: ', trajectory[-1])
print('значение y в точке x: ', f(trajectory[-1]))
```
