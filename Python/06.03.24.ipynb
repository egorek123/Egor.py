```python
import math
from tkinter import *

def func_calculate(operation):
    num1 = int(num_one_entry.get())
    num2 = int(num_two_entry.get())
    if(operation == '+'):
        res = num1 + num2
    elif(operation == '-'):
        res = num1 - num2
    elif(operation == '*'):
        res = num1 * num2
    elif(operation == '/'):
        res = num1 / num2
    elif(operation == '**'):
        res = num1 ** num2
    elif(operation == 'root'):
        res = math.sqrt(num1)
    else:
        res = 'null'

    result.config(text=('Result: ' + str(res)))

    

calculate = Tk()
calculate.title("Calculate")

num_one_text = Label(
    calculate,
    width=34,
    text="Input first num",
    foreground='green',
    background='black'
    )
num_one_text.pack(anchor='w',fill=X)

num_one_entry = Entry(
    calculate,
    width=40
)
num_one_entry.pack(anchor='w',fill=X, pady=10)

num_two_text = Label(
    calculate,
    width=34,
    text="Input second num",
    foreground='green',
    background='black'
    )
num_two_text.pack(anchor='w',fill=X)

num_two_entry = Entry(
    calculate,
    width=40
)
num_two_entry.pack(anchor='w', fill=X, pady=10)

btn_sum = Button(
    calculate,
    text="+",
    width=10,
    command=lambda: func_calculate('+')
    )
btn_sum.pack()

btn_minus = Button(
    calculate,
    text="-",
    width=10,
    command=lambda: func_calculate('-')
    )
btn_minus.pack()

btn_compot = Button(
    calculate,
    text="x",
    width=10,
    command=lambda: func_calculate('*')
    )
btn_compot.pack()

btn_del = Button(
    calculate,
    text=":",
    width=10,
    command=lambda: func_calculate('/')
    )
btn_del.pack()

btn_deg = Button(
    calculate,
    text="^",
    width=10,
    command=lambda: func_calculate('**')
    )
btn_deg.pack()

btn_root = Button(
    calculate,
    text="√",
    width=10,
    command=lambda: func_calculate('root')
    )
btn_root.pack()

result = Label(
    calculate,
    width=34,
    text="Result: ",
    foreground='green',
    background='black'
)
result.pack(anchor='w',fill=X, pady=10)

calculate.mainloop()
```
