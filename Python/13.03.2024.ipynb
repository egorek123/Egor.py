```python
import cv2
import numpy as np

img = cv2.imread("./kot.jpg")

crop_img = img[110:600, 280:500]
# cv2.imshow('1', img)

rotateImg = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
# cv2.imshow('res', crop_img)
# cv2.imshow('1', rotateImg)

hsvImg = cv2.cvtColor(crop_img, cv2.COLOR_RGB2BGR)
# cv2.imshow('hsv', hsvImg)

blurImg = cv2.blur(img, (20,20))
cv2.imshow('blur', blurImg)

kerner = np.array([
    [-1, -1, -1],
    [-1, 9, -1],
    [-1, -1, -1]]
)
snapImg = cv2.filter2D(img, -1, kerner)

cv2.imshow('blur', snapImg)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("./kot.jpg")

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
cv2.imshow('orig', img)
cv2.imshow('sobel x', sobel_x)
cv2.imshow('sobel y', sobel_y)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
