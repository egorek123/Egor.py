```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.metrics import mean_squared_error

boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.2, random_state = 42)
model = DecisionTreeRegressor(max_depth=5)
model.fit(X_train, y_train)
price = model.predict(X_test)

print('MSE: ', mean_squared_error(y_test, price))
print('RMSE: ', sqrt(mean_squared_error(y_test, price)))

print('Цена предсказанная: ', price[0:3])
print('Цена по факту: ', y_test[0:3])
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

depths = []
rase_values = []

for depth in range(1,15):
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X_train, y_train)
    price = model.predict(X_test)
    rase = sqrt(mean_squared_error(y_test, price))
    depths.append(depth)
    rase_values.append(rase)

plt.plot(depths, rase_values)
plt.xlabel('max_depth')
plt.ylabel('RMSE')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.2, random_state = 42)
model = RandomForestRegressor()
model.fit(X_train, y_train)
price = model.predict(X_test)

print('MAE: ', mean_absolute_error(y_test, price) )
print('MSE: ', mean_squared_error(y_test, price))
print('RMSE: ', sqrt(mean_squared_error(y_test, price)))

print('Цена предсказанная: ', price[0:3])
print('Цена по факту: ', y_test[0:3])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import  mean_squared_error
from sklearn.datasets import load_boston

n_estimator_range = range(10, 110, 10)
max_depth_range = range(1,11)
list_rase = []

for n_estimators in n_estimator_range:
    for max_depth in max_depth_range:
        model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth)
        model.fit(X_train,y_train)
        pred = model.predict(X_test)
        nse = mean_squared_error(y_test, pred)
        list_rase.append((n_estimators, max_depth, np.sqrt(nse)))

for n, d, rase in list_rase:
    print(f"n_estimators = {n}, max_depth = {d}, rase = {rase:.2f}")
```
