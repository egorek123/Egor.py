```python
import matplotlib.pyplot as plt

x = [1,4,6,8,10]
y = [-6, 2,1,-5,1]

plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["", "", "", ""]

y = [-20, -25, -10, 5]

plt.bar(x,y,label=' ', color = 'green', alpha = 0.5)
plt.plot(x,y, color='red', marker = '*', markersize = 100)
plt.title(" ")

plt.show()
```

```python
import matplotlib.pyplot as plt

dict_people = {"":30, "":90, "":50}

print(dict_people.values())
print(dict_people.keys())

plt.pie(dict_people.values(), labels=dict_people.keys(), autopct='%1.1f%%')
plt.title(' ')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,4,6,8,10]
y = [-6, 2,1,-5,1]

plt.scatter(x,y, marker = '1', linewidths = 1000)
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,5,1,6])
y = np.array([-10,2,-5,-3,3])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,-y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,5,1,6])
y = np.array([-10,2,-5,-3,3])

plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.bar(-x,y)
plt.subplot(2,3,3)
plt.scatter(x,y)
plt.subplot(2,3,4)
plt.scatter(-x,-y)
plt.subplot(2,3,5)
plt.bar(x,y)
plt.subplot(2,3,6)
plt.plot(-x,-y)
plt.show()
```
