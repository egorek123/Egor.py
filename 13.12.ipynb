```python
import matplotlib.pyplot as mpl
x = ['Декабрь','Январь','Февраль','Март']
y = [-9,-6,-5,8]
mpl.bar(x,y,color='yellow')
mpl.plot(x,y,marker='*',color='red',markersize='100')
mpl.legend
mpl.xlabel('fjj,hg')
mpl.ylabel('hgkg')
mpl.title('oooooo')
mpl.show
```
