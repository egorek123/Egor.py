```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error

x_dt = ege[['Возраст','Количество_решенных_задач','Количество_часов_подготовки','Средний_балл_по_предметам']]
y_dt = ege['Баллы_по_ЕГЭ_по_математике']
x_trnt,x_tstt,y_trnt,y_tstt = train_test_split(x_dt,y_dt,test_size=0.2,random_state=42)
modt = DecisionTreeRegressor(max_depth=5)
modt.fit(x_trnt,y_trnt)
y_rt = modt.predict(x_tstt)
mat = mean_absolute_error(y_pred=y_rt,y_true=y_tstt)
mst = mean_squared_error(y_pred=y_rt,y_true=y_tstt)
print(mat,mst)

import matplotlib.pyplot as plt
plt.scatter(x_tstt['Количество_решенных_задач'],y_tstt)
plt.scatter(x_tstt['Количество_решенных_задач'],y_rt)
plt.show()
```

```python
from sklearn.tree import export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display

name = ['Возраст','Количество_решенных_задач','Количество_часов_подготовки','Средний_балл_по_предметам']
gr = Source(export_graphviz(modt,out_file=None,feature_names=name,filled=True))
display(SVG(gr.pipe(format='svg')))
```
