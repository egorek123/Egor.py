```python
import cv2
import numpy as np

k = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

img = cv2.imread('./mem.jpg')
cr = img[30:100,150:320]
rimg = cv2.resize(img,(0,0),fx=0.5,fy=0.5)
rot = cv2.rotate(rimg,cv2.ROTATE_90_CLOCKWISE)
h = cv2.cvtColor(rot,cv2.COLOR_RGB2BGR)
b = cv2.blur(h,(10,10))
q = cv2.filter2D(b,-1,k)
cv2.imshow('1 window',q)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('./mem.jpg')
sx = cv2.Sobel(q,cv2.CV_64F,1,0,ksize=5)
sy = cv2.Sobel(q,cv2.CV_64F,0,1,ksize=5)
cv2.imshow('1 window',sy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
